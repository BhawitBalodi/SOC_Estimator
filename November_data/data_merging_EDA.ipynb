# Integral SOC

```python
from scipy.integrate import cumulative_trapezoid
```

```python
p = df["Time"][:df.shape[0]]
q = df["Current"][:df.shape[0]]
```

```python
# integral = (cumulative_trapezoid(q,p,initial=0))/(3600*2.6)   # Nominal Capacity - 2.6
integral = (cumulative_trapezoid(q,p,initial=0))  # Nominal Capacity - 2.6
```

```python
df.loc[:, "Integral_soc"] = integral
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(25, 25))
# plt.plot(df['SOC'],color='orange')
plt.plot(df['Integral_soc'],color='green')
plt.ylabel('SOC')
plt.xlabel('Time')
plt.legend(["Soc", "Interal Soc"], loc ="upper left")
plt.title('Dataset')
```
